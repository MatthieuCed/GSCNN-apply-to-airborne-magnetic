### Explanation of other parameters || Explication des autres paramètres

- "**%PCA**" corresponds to a reduction of dimension of the data used perfom the clustering. Principal Component Analysis will be created from the data until *x* percent (corresponding to *%PCA*) of the variance of the data is explained.
- "**Resolution**" allows to reduce the size of the data used to perfom the clustering. For big area, reducing the size will reduce compute time and memory.
- "**Borders**" allows to crop the borders arround the image. As CNN can generate artefacts around borders we recommand to crop borders to remove those artefacts for the clustering step 


- "**%PCA**" correspond à la réduction de dimension des données utilisées par le regroupememnt. Des Analyses du Composant Principal sont calculées, jusqu'à ce que *x* pourcent (correspondant à *%PCA*), de la variance soit expliquée.
- "**Resolution**" permet de réduire la taille des données utilisées. Pour les grandes zones, reduire leur taille permet de réduire la mémoire et le temps requis pour le calcul.
- "**Borders**" permet de couper les bordures des images. Les CNN créant des artefacts autours des bordures, nous conseillons de les enlever pour faciliter le regroupement.
